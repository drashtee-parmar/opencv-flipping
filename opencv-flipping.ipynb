```python
import cv2
from matplotlib import pyplot as plt
```

```python
def plt_imshow(title, image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

```python
image = cv2.imread("opencv_logo.png")
plt_imshow("Opencv Original Image", image)
```

# cv2.flip : flip the images

## flip horizontally by passing value of 1

```python
# flip the image horizontally
print("[INFO] flipping image horizontally...")
flipped = cv2.flip(image, 1)
plt_imshow("Flipped Horizontally", flipped)
```

## flip vertically by passing value of 0

```python
# flip the image vertically
flipped = cv2.flip(image, 0)
print("[INFO] flipping image vertically...")
plt_imshow("Flipped Vertically", flipped)
```

## flip image horizontally and vertically by passing value of -1


```python
# flip the image along both axes
flipped = cv2.flip(image, -1)
print("[INFO] flipping image horizontally and vertically...")
plt_imshow("Flipped Horizontally & Vertically", flipped)
```
